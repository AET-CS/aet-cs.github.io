# Distance Approximators

```python
from matplotlib import pyplot as plt
```

```python
g = 9.81
delta_t = 1/30.0
```

```python
y = 50
v = 0
a = -g
t = 0

y1 = 50
v1 = 0

y_vec = [50]
y1_vec = [50]

for i in range(20000):
    t = t + delta_t
    v = v + a * delta_t
    y = y + v * delta_t

    v1 = v1 + a * delta_t
    y1 = y1 + v1*delta_t + 1/2 * a * delta_t**2 

    y_vec.append(y)
    y1_vec.append(y1)
```

```python
plt.plot(y_vec);
plt.plot(y1_vec);
```
