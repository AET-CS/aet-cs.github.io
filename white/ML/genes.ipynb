```python
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score([0, 0, 1, 1], [0, 0, 1, 1])

normalized_mutual_info_score([0, 0, 1, 1], [1, 1, 0, 0])
```

```python
from sklearn.metrics import mutual_info_score
labels_true = [0, 1, 1, 0, 1, 0]
labels_pred = [0, 1, 0, 0, 1, 1]
mutual_info_score(labels_true, labels_pred)
```

```python
import numpy as np
D = np.random.randint(1,4,(14000,400))
```

```python
import matplotlib.pyplot as plt
```

```python
D.shape
```

```python
def getMI(n):
    for i in range(n):
        for j in range(i,n):
            score = mutual_info_score(D[i],D[j])
```

```python
results = []
from timeit import timeit
```

```python
results = []
for n in range(10):
    results += [(10*n,timeit("getMI(n*10)",globals = globals(), number = 1))]
```

```python
X = [_[0] for _ in results]
Y = [_[1] for _ in results]
plt.plot(X,Y)
```

```python
X2 = np.log(X)[1:]
Y2 = np.log(Y)[1:]
np.polyfit(X2,Y2,deg=1)
```

```python
# Perform a linear fit (degree = 1)
coefficients = np.polyfit(X2, Y2, deg=1)

# Get the slope and intercept
slope, intercept = coefficients

# Generate fitted values
y_fit = slope * X2 + intercept
```

```python
plt.plot(X2,Y2)
plt.plot(X2,y_fit)
```
