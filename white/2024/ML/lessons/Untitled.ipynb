```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
all_r = []
for i in range(100):
    X = np.random.random(100)
    Y = np.random.random(100)
    
    r = (np.cov(X,Y)/np.sqrt((np.var(X)*np.var(Y))))[0,1]
    all_r += [r]
plt.scatter(range(100),all_r)
```

```python
np.mean(all_r), np.std(all_r)
```
