# Distance Approximators

```python
g = 9.81
delta_t = 1/30.0
```

```python
y = 50
v = 0
a = -g
t = 0

y1 = 50
v1 = 50
print(f"{t:.5f} {y:.5f} {y1:.5f}")
for i in range(100):
        t = t + delta_t
        v = v + a * delta_t
        y = y + v * delta_t

        v1 = v1 + a * delta_t
        y1 = y1 + v1*delta_t + 1/2 * a * delta_t**2 
        print(f"{t:.5f} {y:.5f} {v:.5f} {y1:.5f} {v1:.5f}")
```
