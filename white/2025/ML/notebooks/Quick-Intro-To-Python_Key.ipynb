## Matplotlib Samples

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Basic Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Basic Line Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Scatter Plot
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.colorbar(label='Color intensity')
plt.show()
```

```python
# 3. Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 12]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
# 4. Histogram
data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# 5. Subplots
x_sub = np.linspace(0, 10, 100)
y1_sub = np.sin(x_sub)
y2_sub = np.cos(x_sub)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(x_sub, y1_sub, color='red')
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')

axes[1].plot(x_sub, y2_sub, color='purple')
axes[1].set_title('Cosine Wave')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')

plt.tight_layout()
plt.show()
```

```python
# 6. Customizing Styles
plt.style.use('seaborn-v0_8-darkgrid') # Example of using a style

x_style = np.linspace(0, 10, 100)
y_style = np.sin(x_style)

plt.figure(figsize=(8, 6))
plt.plot(x_style, y_style, linestyle='--', color='orange', linewidth=2)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Styled Line Plot')
plt.show()

# You can list available styles with:
# print(plt.style.available)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 7. 3D Plot (Example with a surface plot)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x_3d = np.linspace(-5, 5, 50)
y_3d = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x_3d, y_3d)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')

plt.show()
```
